## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
b_plot = df.boxplot(column = ['Age']) 
b_plot.plot()
```

```python
# your code goes here
b_plot = df.boxplot(column = ['ConvertedComp']) 
b_plot.plot()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
Q1
Q3
IQR = Q3 - Q1
IQR
```

Find out the upper and lower bounds.


```python
# your code goes here
Upper_bound = Q3 + (1.5 * IQR)
Lower_bound = Q1-(1.5 * IQR)
print('lower bounds: ' + str(Lower_bound))

print('upper bounds: ' + str(Upper_bound))
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers = (df["ConvertedComp"] < (Q1 - 1.5 * IQR)) | (df["ConvertedComp"] > (Q3 + 1.5 * IQR))
outliers.value_counts()
```

```python
((df['ConvertedComp'] < (Q1 - 1.5 * IQR)) | (df['ConvertedComp'] > (Q3 + 1.5 * IQR))).sum()
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_filtered = df[~((df["ConvertedComp"] < Lower_bound) | (df["ConvertedComp"] > Upper_bound))]
df_filtered
```

```python
df_filtered["ConvertedComp"].median()
```

```python
df_filtered["ConvertedComp"].mean()
```

```python
df_filtered["Age"].median()
```
